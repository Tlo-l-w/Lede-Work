```python
city_name="Providence" #Question 2
url = f"http://api.weatherapi.com/v1/current.json?key={api_key}&q={city_name}&aqi=no"
response = requests.get(url)
data = response.json()
status_key = data['current']['condition']['text']
print(f"The weather in {city_name} is {data['current']['condition']['text']}")
```

```python
print(f"{city_name} is in {data['location']['country']}") #Question 3
```

```python
feels_like = data['current']['feelslike_f'] #Question 4
current = data['current']['temp_f']
if feels_like > current:
    print(f"It is {current} but feels {(feels_like - current)} degrees warmer at {feels_like}")
elif feels_like == current:
    print(f"It is {current} and OOWEEEE does it feel like it")
else:
    print(f"It is {current} but feels {(current - feels_like)} degrees colder at {feels_like}")
```
